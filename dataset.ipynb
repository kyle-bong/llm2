```python
import pandas as pd
import os
import json
```

```python

def merge_json(path):
    file_list = []
    total_df = pd.DataFrame()
    for root, dirs, files in os.walk(path):
        for dir in dirs:
            try:
                for file in os.listdir(os.path.join(path, dir)):
                    if file.endswith('.json'):
                        with open(
                            os.path.join(path, dir, file), 'r'
                        ) as f:
                            df = json.loads(f.read())
                            df = pd.json_normalize(df)
                            # 필요한 정보만 추출
                            # df = df[['info.essay_id', 'info.essay_prompt', 'paragraph', 'score.essay_scoreT_avg',  'student.student_grade']]
                            total_df = pd.concat([total_df, df])
            except:
                pass
            
    return total_df
```

```python
import pandas as pd
train_df = pd.read_json('dataset/total_train_dataset.json')
val_df = pd.read_json('dataset/total_val_dataset.json')
```

```python
import re
def make_prompt(essay_prompt, paragraph, score, grade, correction):
    result = ''
    if grade:
        result += f'### 학년: {grade.strip()}\n\n' 
    if essay_prompt:
        essay_prompt = essay_prompt.replace('\n', ' ')
        essay_prompt = re.sub(r'\s+', ' ', essay_prompt)
        result += f"### 문제: {essay_prompt.strip()}\n\n"
    if paragraph:
        result += f'### 답안: {paragraph.strip()}\n\n'
    if score:
        result += f'### 점수: {str(round(score, 2)).strip()}\n\n'
    if correction:
        result += f'### 교정 내용: \n{correction}'
    result += '<|endoftext|>'
    return result
```

```python
from collections import defaultdict
train_txt = defaultdict(list)
for i in range(len(train_df)):
    prompt = make_prompt(train_df['info.essay_prompt'].iloc[i],
                                 train_df['paragraph'].iloc[i],
                                 train_df['score.essay_scoreT_avg'].iloc[i],
                                 train_df['student.student_grade'].iloc[i],
                                 train_df['squeeze_correction'].iloc[i]
                                )
    if prompt:
        train_txt['text'].append(prompt)
```
